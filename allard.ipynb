# Light Visibility Distance Calculator

This calculator takes in a value for the light illuminance at a given distance away from a point source of light, and calculates the distance from which the light will still be visible to the naked eye, were it to continue travelling away from the light source.
It does not calculate the distance from the light source itself, only the distance between the given starting point and where the light becomes too dim to see.

The calculator uses Allards Law for light illuminance at a distance, and iterates using the Newton-Raphson method to solve for the distance.

## Parameters

 - E - The light illuminance at the starting point, measured in lux
 - d - The distance from the light source to the starting point, measured in nautical miles

The light illuminence at the starting point and starting distance must be provided, the rest of the parameters are optional. If optional parameters are not given, the calculator uses default parameters.

Optional parameters:

 - E_threshold - Illuminance threshold for naked eye visibility, measured in lux
 - T - Atmospheric transmission factor, unitless (assumed constant)
 - tol - Solver error tolerance (should be small for accurate distances)



```python
import math

def calculateVisibleDistance(E, d, E_threshold = 0.05, T = 0.74, tol = 1e-9):

    # Conversion from nautical miles to meters
    d = d*1852
    
    # Function definitions
    def allards_law_for_illuminance_null_form(r):
        return (E*(d/r)**2*T**(r-d)) - E_threshold
        
    def deriv_allards_law_for_illuminance_null_form(r):
        return (E*d**2*T**(r-d)*(r*math.log(T)-2))/r**3

    # Newton-Raphson
    r = d + 1    
    while abs(allards_law_for_illuminance_null_form(r)) > tol:
        r = r - allards_law_for_illuminance_null_form(r)/deriv_allards_law_for_illuminance_null_form(r)

    # We want distance from starting point, not light source
    x = r - d
    
    # Conversion from meters to nautical miles
    x = x/1852
    
    return x
    
```

```python
# Enter values for E and d here
E = 0.2
d = 8

x = calculateVisibleDistance(E, d) # Add optional parameters if desired

print("If the light measures", E, "lux at a distance of", d, "nautical miles from the light source, it will still be visible", f"{x:.4f}", "nautical miles further away")
```
